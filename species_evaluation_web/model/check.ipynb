```python
import joblib, pandas as pd

model = joblib.load("model.joblib")

a = {f"S{i}": 6 for i in range(1, 21)}
b = {f"S{i}": 6 for i in range(1, 21)}
b["S1"] = 100   # big change

pa = float(model.predict(pd.DataFrame([a]))[0])
pb = float(model.predict(pd.DataFrame([b]))[0])

print("all 6:", pa)
print("S1=100:", pb)
print("diff:", pb - pa)
```
